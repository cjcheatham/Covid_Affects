```python
# import dependencies
import pandas as pd
import glob
import matplotlib.pyplot as plt
import numpy as np 
from pandas_geojson import to_geojson
```

```python
# load in data
cases_data = pd.read_csv("data/cleaned/confirmed_covid_cases_cleaned.csv")
cases_data
```

```python
# check data types
cases_data.info()
```

```python
# rearrange in a df
df = pd.DataFrame({
    "state_id" : cases_data["state_id"],
    "state_abv" : cases_data["state_abv"],
    "state" : cases_data["state"],
    "date" : cases_data["date"],
    "cases" : cases_data["cases"],
    "population" : cases_data["population"],
    "cases_per_cap" : cases_data["cases_per_cap"]
})
df
```

```python
# convert date to datetime format
df["date"] = pd.to_datetime(df["date"])
df.info()
```

```python
# export to a json file
df.to_geojson("data/cleaned/cases.geojson")
```

```python
# drop 2022 and 2023
df2 = df[(df['date'].dt.year < 2022) | (df['date'].dt.year > 2023)]
```

```python
df2
```

```python
df2.to_csv("data/cleaned/confirmed_covid_cases_2020_2021.csv")
df2.to_json("data/cleaned/cases_2020_2021.json")
```
