```python
#simplifying dataset
selected_columns = ['state_name', '2_letter_abb', 'year', 'month_name', 'day', 'spend_all']

state_wide_perchng = state_daily[selected_columns].copy()
```

```python
state_wide_perchng
```

```python
state_wide_perchng.dtypes
```

```python
state_wide_perchng['spend_all'] = pd.to_numeric(state_wide_perchng['spend_all'], errors='coerce')

# Calculate cumulative percentage change
state_wide_perchng['cumulative_percent_change'] = (
    state_wide_perchng.groupby(['state_name', 'year'])['spend_all']
    .pct_change()
    .add(1)
    .cumprod()
    .sub(1)
    .mul(100)
)

# Drop NaN values resulting from the first row of each group
state_wide_perchng.dropna(subset=['cumulative_percent_change'], inplace=True)
```
