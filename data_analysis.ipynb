```python
#loading the CSV files into a dictionary of dataframes

dataframes = {} #empty dictionary to hold dfs

for file_name, file_path in file_paths.items():
    try:
        df = pd.read_csv(file_path, low_memory=False) #loading the CSV file into a Dataframe
        dataframes[file_name] = df #storing dataframe in dictionary
        
        #displaying the first few rows of each df
        print(f"Preview of {file_name}:")
        display(df.head())
        print("\n")
    
    except Exception as e:
        print(f"Error loading P{file_name}: {e}")
```
