# Created bar chart to display percentage change in total consumer spending during Covid-19

```
grouped_national_df['Date'] = grouped_national_df['year'].astype(str) + '-' + grouped_national_df['month'].astype(str)

#creating the bar chart for the "Total Spending" column
national_spend_change = px.bar(grouped_national_df, x='Date', y='Total Spending', 
             labels={'Total Spending': 'Total Spending'},
             title='Total Spending During Covid-19')

#updating the x-axis labels for better readability
national_spend_change.update_layout(xaxis_title='Month-Year')
national_spend_change.update_layout(title=dict(x=0.5, y=0.95, xanchor='center', yanchor='top'))

#increasing font size
national_spend_change.update_layout(font=dict(size=16))

#display the chart
national_spend_change.show()
```

```
#exporting the chart as HTML
national_spend_change.write_html("National Spending % Change.html")
```

# Creating a scatter plot with dropdown of variables to display an overview of consumer spending during Covid-19 (2020-2023)

```
import plotly.graph_objects as go

#creating a new column for Date
grouped_national_df['Date'] = grouped_national_df['year'].astype(str) + '-' + grouped_national_df['month'].astype(str)

#getting the columns with names starting with 'Spending....'
spend_columns = [col for col in grouped_national_df.columns if col.startswith('Spending')]

#creating the initial scatter plot
overview_scatter_fig = go.Figure()

for col in spend_columns:
    overview_scatter_fig.add_trace(go.Scatter(x=grouped_national_df['Date'], y=grouped_national_df[col], mode='markers',
                                     name=col, hoverinfo='y+name'))

#updating layout
overview_scatter_fig.update_layout(title='Overview of Consumer Spending by Category During Covid-19',
                          xaxis_title='Month-Year', yaxis_title='Percent Change')

#adding dropdown menu for variable selection
buttons = [
    dict(label='All',
         method='update',
         args=[{'visible': [True] * len(scatter_fig.data)},
               {'title': 'Overview of Consumer Spending by Category During Covid-19',
                'xaxis_title': 'Month-Year',
                'yaxis_title': 'Spending'}])
] + [
    dict(label=col,
         method='update',
         args=[{'visible': [col == trace.name for trace in scatter_fig.data]},
               {'title': f'Percent Change for {col}',
                'xaxis_title': 'Month-Year',
                'yaxis_title': 'Percent Change'}]) for col in spend_columns
]

overview_scatter_fig.update_layout(updatemenus=[dict(type='dropdown',
                                             showactive=True,
                                             active=0,
                                             buttons=buttons,
                                             x=0.1,  # Adjust the x position of the dropdown
                                             y=1.05,  # Adjust the y position of the dropdown
                                             xanchor='left',
                                             yanchor='top',
                                             bgcolor='white',
                                             bordercolor='gray',
                                             font=dict(color='black'))])

#adjusting the chart size
overview_scatter_fig.update_layout(height=800, width=950)

#displaying the initial scatter plot
overview_scatter_fig.show()
```

```
#exporting the chart as HTML
overview_scatter_fig.write_html("Overview_SpendingbyCategory.html")
```
